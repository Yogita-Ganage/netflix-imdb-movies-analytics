####Number of titles released per year

```python
%sql
SELECT
  releaseYear,
  COUNT(*) AS titleCount
FROM gold.movies
GROUP BY releaseYear
ORDER BY releaseYear,count(*) DESC;
```

####Top directors by content production

```python
%sql
SELECT
  director,
  COUNT(*)        AS titleCount,
  ROUND(AVG(imdbRating), 2) AS avgImdbRating
FROM gold.movies
WHERE director IS NOT NULL
GROUP BY director
ORDER BY titleCount DESC, avgImdbRating DESC
LIMIT 20;
```

####Most common age certifications

```python
%sql
SELECT
  ageCertification,
  COUNT(*) AS titleCount
FROM gold.movies
GROUP BY ageCertification
ORDER BY titleCount DESC;
```
